```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df=pd.read_csv('daily_data.csv')
df
```

```python
fig = go.Figure(data=[go.Candlestick(
                x=df['datetime'],
                open=df['open'],
                high=df['high'],
                low=df['low'],
                close=df['close'])])

fig.show()
```

```python
def ifhammer(candle):
  if candle['high'] - candle['open'] == 0:
        return 0
  ratio=(candle['open']-candle['low'])/(candle['high']-candle['open'])
  if(ratio>2):
    return 1
  else: return 0

```

```python
hammers=[]
hammers=df.apply(ifhammer, axis=1)
ans=hammers.sum()
print(ans)
```
