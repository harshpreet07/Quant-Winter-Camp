```python
import pandas as pd
df = pd.read_csv('daily_data.csv')
df
```

```python
def solve(df, take_profit_percent, stop_loss_percent):
    buy_list = []
    sell_list = []
    in_trade = False
    for i in range(len(df)):
        if not in_trade:
          #agar last profit mai huya hoga
            if df['close'][i] >= df['close'][0] * (1 + take_profit_percent):
                in_trade = True
                buy_list.append(i)
          #agar last loss huya
            elif df['close'][i] <= df['close'][0] * (1 - stop_loss_percent):
                in_trade = True
                buy_list.append(i)
        else:
            if df['close'][i] <= df['close'][buy_list[-1]] * (1 - stop_loss_percent):
                in_trade = False
                sell_list.append(i)
            elif df['close'][i] >= df['close'][buy_list[-1]] * (1 + take_profit_percent):
                in_trade = False
                sell_list.append(i)
    print("Buy list:", buy_list)
    print("Sell list:", sell_list)
```

```python
solve(df, take_profit_percent=0.04, stop_loss_percent=0.02)
```
